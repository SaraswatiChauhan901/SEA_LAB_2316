## 5. Distribution of Case Fatality Rate (CFR)

The histogram shows how the case fatality rate (deaths per 100 confirmed cases) is distributed across all nations in your dataset. Most countries cluster at a lower CFR, likely under 5%, but there are always outlier nations with higher rates. These extremes may stem from overwhelmed health systems, populations at higher risk (e.g., older adults), limited access to care, or challenges in accurate case detection.

**Analysis:**
- Variability in CFR emphasizes the importance of targeted responses to vulnerable populations and regions.
- High CFR outliers suggest further investigation into healthcare resource allocation, disease monitoring, and broader pandemic preparedness.
